```python
features = movies[["id", "original_title", "genres", "overview"]]
features
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10866, stop_words="english")
```

```python
cv
```

```python
vector = cv.fit_transform(features["tags"].values.astype("U")).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
features.info()
```

```python
similar_movies = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector: vector[1])
result = []
for similar_movie in similar_movies[:5]:
    result.append(similar_movie)
    print(features.iloc[similar_movie[0]].original_title)

#result
```

```python
def recommend(movies):
    index = features[features["original_title"] == movies].index[0]
    similar_movies = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])

    for similar_movie in similar_movies[:5]:
        print(features.iloc[similar_movie[0]].original_title)
```

```python
recommend("Insurgent")
```
